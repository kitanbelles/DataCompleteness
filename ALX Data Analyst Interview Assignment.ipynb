```python
# Reading in the necessary library 
import pandas as pd 
```

```python
# Reading in the data as a Dataframe 
df = pd.read_csv("DA Test Data.csv")
```

```python
# Exploring the data 
df.columns
```

```python
# Exploring the data
df.shape
```

```python
# Counting number of columns in each row with values 
# Getting the completeness valus and writing it into the dataframe 
for index, row in df.iterrows():
    count = 0; 
    if(pd.isnull(row["ID"]) or row["ID"] == ""):
        count+=1;
    if(pd.isnull(row["Source"]) or row["Source"] == ""):
        count+=1;
    if(pd.isnull(row["Customer Segment"]) or row["Customer Segment"] == ""):
        count+=1;
    if(pd.isnull(row["Membership Type"]) or row["Membership Type"] == ""):
        count+=1;
    if(pd.isnull(row["Membership Level"]) or row["Membership Level"] == ""):
        count+=1;
    if(pd.isnull(row["Affiliation"]) or row["Affiliation"] == ""):
        count+=1;
    if(pd.isnull(row["IsActive"]) or row["IsActive"] == ""):
        count+=1;
    if(pd.isnull(row["Gender"]) or row["Gender"] == ""):
        count+=1;
    if(pd.isnull(row["Citizenship"]) or row["Citizenship"] == ""):
        count+=1;
    if(pd.isnull(row["Age"]) or row["Age"] == ""):
        count+=1;
    if(pd.isnull(row["Lifetime Days"]) or row["Lifetime Days"] == ""):
        count+=1;
    if(pd.isnull(row["Has Linkedin"]) or row["Has Linkedin"] == ""):
        count+=1;
    if(pd.isnull(row["City"]) or row["City"] == ""):
        count+=1;
    if(pd.isnull(row["Country"]) or row["Country"] == ""):
        count+=1;
    if(pd.isnull(row["Industry"]) or row["Industry"] == ""):
        count+=1;
    if(pd.isnull(row["Interviewed"]) or row["Interviewed"] == ""):
        count+=1;
    if(pd.isnull(row["Pathfinder Assigned"]) or row["Pathfinder Assigned"] == ""):
        count+=1;
    if(pd.isnull(row["Customer Call Status"]) or row["Customer Call Status"] == ""):
        count+=1;
    if(pd.isnull(row["Languages"]) or row["Languages"] == ""):
        count+=1;
    completeness = (count/19) *100
    #print(completeness)
    df.at[index,"Completeness"] = completeness 
    
```

```python
# Checking the dataframe after inserting the completeness column 
df.sample(10)
```

```python
# Writing data back to csv 
df.to_csv("DA Test Data Complete.csv", index= False)
```
